```python
#sujet0

#grosses variables globales de porc
currentnum=0
end_reached=0
url_start="https://forum.hardware.fr/forum2.php?post="
url_mid="&cat=13&config=hfr.inc&cache=&page=1&sondage=0&owntopic=0&word=&spseudo="+pseudo+"&firstnum=0&"
url_end="&filter=1"
```

```python
response1=requests.get(url_start+str(21512)+url_mid+"currentnum="+str(0)+url_end, headers=headers)
soup=bs4.BeautifulSoup(response1.text,'html.parser')
```

```python
dates=[]
subjects=[]
texts=[]
```

```python
def get_values(subject_var,currentnum_var): #fonction qui va dans une page d'un sujet
    
    global url_start
    global url_mid
    global url_end
    global currentnum
    global dates
    global end_reached

    
    
    response1=requests.get(url_start+str(subject_var)+url_mid+"currentnum="+str(currentnum_var)+url_end, headers=headers)
    soup=bs4.BeautifulSoup(response1.text,'html.parser')
    div_box = soup.find_all("div", {"class":"left"})
    
    messages_box=soup.find_all("table", {"class":'messagetable'})
    if len(messages_box)>0:
        messages_box.pop(1)
    
    if len(div_box)<=1: #condition pas fin de liste
        end_reached=1

    else:
        currentnum=div_box[0].find('input', {'id': 'currentnum'}).get('value') #mise à jour pour suivant
        for event in range(1,len(messages_box)-1):
            m=messages_box[event].find("td", {"class":'messCase2'}).find_all("div", recursive=False)[1].find_all("p", recursive=False)
            msg=''
            for elem in range(0,len(m)):
                msg=msg+m[elem].text
            prov=messages_box[event].find("td", {"class":'messCase2'}).div.div.text
            match = re.search(r'\d{2}-\d{2}-\d{4}', prov)
            date = datetime.datetime.strptime(match.group(), '%d-%m-%Y').date()
            dates.append(date)
            subjects.append(subject_var)
            texts.append(msg)
```

```python
for subject_number in subjects_number:
    end_reached=0
    currentnum=0
    #print(subject_number)
    while end_reached==0:
        #print(currentnum)
        get_values(subject_number,currentnum)
```

```python
import pandas as pd

df = pd.DataFrame({'subjects': subjects,
'date': dates,
'text': texts,
                  })
```
