```python
url="https://www.seloger.com/list.htm?projects=2&types=1%2C2&natures=1&places=%5B%7Bcp%3A75%7D%5D&enterprise=0&qsVersion=1.0&LISTING-LISTpg="
headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36',}

price_tag="Price__Label-sc-1g9fitq-1 lfyHpt"
price_list=[]

elements_tag = "ContentZone__Tags-wghbmy-7 fNlgGF"
element_list=[]

location_tag = "Card__LabelGap-sc-7insep-5 ContentZone__Address-wghbmy-2 eAcNsS"
location_list=[]

reference_list=[]

for i in range(1,nb_pages+1):
    url_page=url+str(i)
    response_page = requests.get(url_page, headers=headers)
    page_soup=bs4.BeautifulSoup(response_page.text,'html.parser')
    
    price_box=page_soup.find_all("div", {"class":price_tag})
    for price in price_box:
        price_list.append(price.findAll(string=True))
    
    elements_box=page_soup.find_all("ul", {"class":elements_tag})
    for element in elements_box:
        element_list.append(element.find_all(string=True))
    
    location_box=page_soup.find_all("div", {"class":location_tag})
    for element in location_box:
        location_list.append(element.find_all(string=True))
    
    reference_box=page_soup.find_all("a", {"z":1})
    for element in reference_box:
        reference_list.append(element.get('href'))
    
    print("Page" + str(i))
    sleep(randint(1,5)) #pour simuler scroll page
```

```python
reference_list=reference_list[::2] #valeurs en doublon
```

# Data cleaning

```python
df["Pieces"] = df["Concat"].str.extract('(\d) p')
df["Chambres"] = df["Concat"].str.extract('(\d) ch')
df["Superficie"] = df["Concat"].str.extract('(\d*\,\d*|\d*) m²')
df["Arr"] = df["Arrondissement"].str.extract('(\d*)ème')
```

```python
df["Prix"]=df.Prix.str.replace('€','')
df["Prix"]=df.Prix.str.replace('\xa0','') #espace relou
df["Superficie"]=df.Superficie.str.replace(',','.').astype(float)
df["Num_annonce"]=df["URL"].str.extract('(\d*).htm?')
```

```python
#on retire les viagers
to_drop=["bouquet"]
df=df[~df['Prix'].isin(to_drop)]

#On retire les annonces sans prix ou superficie
df = df.dropna(subset=['Prix', 'Superficie', 'Pieces','Arr'])
df = df.reset_index(drop=True)
```

```python
df_clean=df[["Num_annonce","Superficie","Pieces","Prix","Arr","Quartier","URL"]].copy()
```
